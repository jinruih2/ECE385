```python
import os
all_content = []
for filename in filenames:
    with open(f"sprite_bytes/{filename}") as f:
        content = f.read()
        try:
            all_content += [f'{_}' for _ in content.strip('\n').split('\n')]
        except:
            print(content)
with open('all.txt', 'w') as f:
    for _ in all_content:
        f.write(f'{_}\n')
```
